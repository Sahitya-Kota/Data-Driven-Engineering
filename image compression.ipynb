# Image Compression using SVD

```python
import numpy as np 
import matplotlib.pyplot as plt
import os
from matplotlib.image import imread
```

```python
plt.rcParams['figure.figsize'] = [16,8]

img = imread('.images/100_0181.JPG')
X = np.mean(img, -1) # RGB -> grayscale
```

```python
img = plt.imshow(X)
img.set_cmap('gray')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X,full_matrices = False)
S = np.diag(S)
```

```python
j = 0
for r in (5,20,100):
    Xapprox = U[:,:r] @ S[:r,:r] @ VT[:r, :]
    plt.figure(j+1)
    j+=1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.title(f'r={r}')
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Value') #log(Sj) vs j
plt.show()

plt.figure(2)
plt.semilogy(np.cumsum(np.diag(S))/np.sum(np.diag(S)))  # 1->r sum(Sj)/ 1->m sum(Sj)
plt.title('Singular Values: Cumultive Sum')
plt.show()
```
